```python
#Tokenization
from indicnlp.tokenize import indic_tokenize
def tokenization(indic_string):
    tokens = []
    for t in indic_tokenize.trivial_tokenize(indic_string):
        tokens.append(t)
    return tokens
data['original'] = data['original'].apply(lambda x: tokenization(x))
```

```python
import random
from sklearn.model_selection import train_test_split

X = data['original'].tolist()
y = data['nc'].tolist()

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)

# Split the train set further into train and validation sets
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size=0.15, random_state=0)

train_percentage = len(X_train) / len(data) * 100
validate_percentage = len(X_validate) / len(data) * 100
test_percentage = len(X_test) / len(data) * 100

print("Train set:", train_percentage, "%")
print("Validate set:", validate_percentage, "%")
print("Test set:", test_percentage, "%")
```
